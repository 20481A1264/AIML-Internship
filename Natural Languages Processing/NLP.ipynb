# Import The Libraries

```python
import pandas as pd
import nltk # spacy
import re # removing the special characters
```

```python
nltk.download('stopwords')
 
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()

data=[]
```

# Download the dataset

Link : https://drive.google.com/file/d/1LojCB1VAeN8RP1yNz4mQ5F4IKVs4evUZ/view?usp=sharing

```python
#2.Read the dataset
dataset=pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t')
```

# Text Preprocessing

```python
#3.Text Preprocessing:

for i in range(0,1000):
    review=dataset['Review'][i]
    
    #i)remove the unnecessary things
   
    #Wow... Loved this place.
    review=re.sub('[^a-zA-Z]',' ',review)
   
    #Wow    Loved this place 
    
    #ii)lower case the text
    
    #Wow  wow
    review=review.lower()
    
    #wow    loved this place 
    
    #iii)split the text
    
    review=review.split()
    
    #['wow','loved','this','place']
    
    #iv)stemming
    #v)remove stopwords
    
    review=[ps.stem(word) for word in review if not word in 
             set(stopwords.words('english'))]
    
    #['wow','love','place']
    review=' '.join(review)
    #wow love place
    data.append(review)
    
    
```

```python
#4 tokenization or vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000)
```

```python
cv
```

```python
# vii)spliting the datat into x,y
x=cv.fit_transform(data).toarray() 
y=dataset.iloc[:,1:2].values
```

# Train,Test & Split

```python
#viii)split the train and test
    
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# ANN Layers

```python
#4. ANN layers
 
#import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

# Model Building

```python
model=Sequential()
```

```python
model.add(Dense(1565,activation='relu'))
model.add(Dense(3000,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10)
```

# Save The Model

```python
#5. save our model

model.save('restaurant.h5')
```
